```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt


```

```python
# Load the dataset
df1=pd.read_csv("C:/Users/aman/Desktop/BITS Research/Data analysis/copy- LoRaWAN dataset.csv")

# Drop non-numeric columns (if any) and target column (if applicable)
df=df1.drop(['index','timestamp','device_id','sf'],axis='columns')
```

```python
# Standardize the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# Apply PCA
pca = PCA()
pca.fit(df_scaled)

# Explained variance
explained_variance = pca.explained_variance_ratio_

# Plot explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance.cumsum(), marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA - Explained Variance')
plt.grid()
plt.show()

# Components and feature importance
components = pca.components_
df_components = pd.DataFrame(components, columns=df.columns)

print("Principal Components:\n", df_components)
print("Explained Variance Ratio:\n", explained_variance)
```
