```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data based on the image provided
data = {
    'Category': ['RSSI', 'SNR', 'Frequency', 'Height', 'Distance'],
    'Series1': [57.8106, 54.3645, 44.6041, 52.5358, 53.1203],
    'Series2': [52.0416, 52.8281, 52.5337, 52.5965, 52.5337],
    'Series3': [59.4378, 56.9425, 52.5337, 52.9371, 52.9371],
    'Series4': [59.4378, 56.9425, 52.5337, 52.9371, 52.9371],
    'Series5': [57.18195, 55.2694, 50.5513, 52.751625, 52.88205]
}

# Convert data to a DataFrame
df = pd.DataFrame(data)

# Melt the DataFrame to have a long-form format suitable for Seaborn
df_melted = df.melt(id_vars='Category', var_name='Series', value_name='Value')

# Plotting
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")
plt.gca().set_facecolor('#EAEAF2')

# Using a more refined palette
palette = sns.color_palette("coolwarm", 5)
barplot = sns.barplot(x='Value', y='Category', hue='Series', data=df_melted, palette=palette, edgecolor=".2")

# Customizing the plot
barplot.set_title('Model Accuracies', fontsize=20, weight='bold', pad=20)
barplot.set_xlabel('Accuracy %', fontsize=16, labelpad=10)
barplot.set_ylabel('Category', fontsize=16, labelpad=10)
barplot.set_xlim(40, 70)  # Set x-axis limits from 30 to 70

# Add grid lines for better readability
barplot.grid(True, linestyle='--', linewidth=0.5, color='gray', axis='x', zorder=0)

# Display values on bars for better insight
for index, row in df_melted.iterrows():
    barplot.text(row['Value'] + 0.5, index, f'{row["Value"]:.1f}', color='black', ha="center", va='center')

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Series', title_fontsize='13', fontsize='11', loc='upper right', frameon=False)

# Removing spines for a cleaner look
sns.despine(left=True, bottom=True)

plt.tight_layout()
plt.show()
```
